```python
import numpy as np
from numpy import linalg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x,y):
    return .5*x**2 + 3.5*y**2
def g(x,y):
    return -x+y-1
```

```python
def descente_gradient_pas_fixe_quadratique(A, b, X0, taux = 0, e = 1e-4):
    vps=np.linalg.eigvals(A)
    lamda1=vps.min(axis=0) #valeur propre minimal de A
    lamdad=vps.max(axis=0) #valeur propre maximal de A
    if taux == 0: #si le pas n'est pas donné on ulilise le meilleur pas fixe
        taux=2/(lamda1 + lamdad)
    dist = 1 # distance ||Xn+1 - Xn||
    Xn = X0
    LX = [X0[0]]
    LY = [X0[1]]
    nbIt = 0
    while dist > e:
        Y  = Xn - taux*(A@Xn - b)
        LX.append(Y[0])
        LY.append(Y[1])
        dist = np.linalg.norm(Y-Xn)
        Xn = Y
        nbIt+=1
    return np.array(LX), np.array(LY), nbIt   
```

```python
def taux_opt(X):
    return ( X[0]**2 + 49*X[1]**2 )/(X[0]**2 + 343*X[1]**2)

def descente_gradient_pas_optimal(A, b, X0, e = 1e-4): 
    dist = 1 # distance ||Xn+1 - Xn||
    Xn = X0
    LX = [X0[0]]
    LY = [X0[1]]
    nbIt = 0
    while dist > e:
        gradient_f = A@Xn - b
        if (gradient_f.all() == 0):
            break
        taux_n = taux_opt(Xn)
        Y  = Xn - taux_n*(gradient_f)
        LX.append(Y[0])
        LY.append(Y[1])
        dist = np.linalg.norm(Y-Xn)
        Xn = Y
        nbIt+=1
    return np.array(LX), np.array(LY), nbIt 
```

```python
def descente_gradient_acceleree_Nesterov(A, b, X0, X_1, e):
    vps=np.linalg.eigvals(A)
    m=vps.min(axis=0)
    M=vps.max(axis=0)
    t = 1/M
    beta = (M**(.5) - m**(.5))/(M**(.5) + m**(.5))
    dist = 1
    Xn = X0
    Xm = X_1
    LX = [X0[0]]
    LY = [X0[1]]
    nbIt = 0
    while dist > e:
        Y = beta*(Xn - Xm)
        Xk = Xn - t*(A@(Xn - Y) - b) + Y
        LX.append(Xk[0])
        LY.append(Xk[1])
        dist = np.linalg.norm(Xk-Xn)
        Xm = Xn
        Xn = Xk
        nbIt+=1
    return np.array(LX), np.array(LY), nbIt
```

```python
def Plot(X, Y, f, forms, labels, title):
    x1 = np.linspace(-4, 10, 150)
    x2 = np.linspace(-4.5, 4.5, 150)
    # x1 = np.linspace(-1.5, 4.5, 150)
    # x2 = np.linspace(-1, 4.5, 150)
    X1, X2 = np.meshgrid(x1, x2)
    Z = f(X1, X2)
    fig = plt.figure(figsize = (10,7))
    contours = plt.contour(X1, X2, Z, 20)
    plt.clabel(contours, inline = True, fontsize = 10)
    plt.title(title, fontsize=15)
    for x, y , form, label in zip(X, Y, forms, labels):
        plt.plot(x, y)
        plt.plot(x, y, form, label=label)

    plt.xlabel('x', fontsize=11)
    plt.ylabel('y', fontsize=11)
    plt.colorbar()
    plt.legend(loc = "upper left")
    plt.show()
```

```python
A = np.array([[1,.0],[.0,7]])
b = 0
X0 = np.array([7,1.5])
taux = 1/7
e = 1e-3
```

```python
X_fix, Y_fix, n_fix = descente_gradient_pas_fixe_quadratique(A, b, X0, taux, e)
X_opt, Y_opt, n_opt = descente_gradient_pas_optimal(A, b, X0, e )
X_1 = np.zeros(2)
X_acc_Nes, Y_acc_Nes, n_acc_Nes = descente_gradient_acceleree_Nesterov(A, b, X0, X_1, e)
```

```python
XX = [X_fix,X_opt,X_acc_Nes]
YY = [Y_fix,Y_opt,Y_acc_Nes]
forms = ['*','x','.']
labels = ['fix path','optimal path','gradient accéléré de Nesterov']
print('Le nombre d''itérations prise par la méthode de descente de gradient à pas fixe : ',n_fix)
print('Le minimun trouvé est :',X_fix[-1], Y_fix[-1])
print('Le nombre d''itérations prise par la méthode de descente de gradient à pas optimal : ',n_opt)
print('Le minimun trouvé est :',X_opt[-1], Y_opt[-1])
print('Le nombre d''itérations prise par la méthode de descente de gradient accéléré de Nesterov : ',n_acc_Nes)
print('Le minimun trouvé est :',X_acc_Nes[-1], Y_acc_Nes[-1])

Plot(XX, YY, f,forms, labels, "Evolution of the f function during gradient descent")
```

```python
def projection(X):
    # u = np.array([1,1])
    # return ((X-np.array([0,1]))@u/np.linalg.norm(u)**2)*u + np.array([0,1])
    return np.array([(X[0]+X[1]-1)/2,(X[0]+X[1]+1)/2])
   
def gradient_projete(A, b, X0, e = 1e-4): 
    dist = 1
    Xn = X0
    LX = [X0[0]]
    LY = [X0[1]]
    nbIt = 0
    while dist > e:
        gradient_f = A@Xn - b
        if (gradient_f.all() == 0):
            break
        taux_n = taux_opt(Xn)
        Y  = projection(Xn - taux_n*gradient_f)
        LX.append(Y[0])
        LY.append(Y[1])
        dist = np.linalg.norm(Y-Xn)
        Xn = Y
        nbIt+=1
    return np.array(LX), np.array(LY), nbIt 
```

```python
X0 = np.array([4,5.5])
X_pr, Y_pr, n_pr = gradient_projete(A, b, X0, e)
```

```python
XX = [X_pr]
YY = [Y_pr]
forms = ['*']
labels = ['projection']
print('Le nombre d''itérations prise par la méthode de descente de gradient à pas optimal avec projection : ',n_pr)
print('Le minimun trouvé est :',X_pr[-1], Y_pr[-1])
Plot(XX, YY, f,forms, labels, "Evolution of the f function during gradient descent with projection")
```

```python
def penality(X):
    return g(X[0],X[1])**2
def Penalisation_exterieure(X0,e):
    k = 1
    m = 1
    dist = 1
    LX = [X0[0]]
    LY = [X0[1]]
    while dist > e:
        A = np.array([[1+2*m,-2*m],[-2*m,7+2*m]])
        b = np.array([-2*m,2*m]) # c = m
        X, Y , n = descente_gradient_pas_fixe_quadratique(A, b, X0, taux = 0, e = 1e-3)
        Xk = np.array([X[-1],Y[-1]]) # Xk
        LX.append(Xk[0])
        LY.append(Xk[1])
        dist = np.linalg.norm(Xk-X0) # distance ||Xk - Xk-1||
        X0 = Xk # Xk sera la point d'initialisation de la prochaine itération
        m+=1
        k+=1
    return np.array(LX), np.array(LY), (k-1)
        
```

```python
X_pen, Y_pen, n_pen = Penalisation_exterieure(X0,e)
```

```python
XX = [X_pen]
YY = [Y_pen]
forms = ['.']
labels = ['pénalisation extérieure']
print('Le nombre d''itérations prise par la méthode de pénalisation : ',n_pen)
print('Le minimun trouvé est :',X_pen[-1], Y_pen[-1])
Plot(XX, YY, f,forms, labels, "Evolution of the f function during penalisation method")
```

```python
def Uzawa(X0, l0, t, e):
    dist = 1
    LX = [X0[0]]
    LY = [X0[1]]
    l = l0
    Xn = X0
    nbIt = 0
    while dist > e:
        A = np.array([[1,0],[0,7]])
        b = np.array([l,-l]) # c = -l
        X, Y , n = descente_gradient_pas_fixe_quadratique(A, b, Xn, taux = 0, e = 1e-3)
        l = max(0,l+t*g(X[-1],Y[-1]))
        Xk = np.array([X[-1],Y[-1]])
        LX.append(Xk[0])
        LY.append(Xk[1])
        dist = np.linalg.norm(Xk-Xn) # distance ||Xk - Xk-1||
        Xn = Xk
        nbIt+=1
    return np.array(LX), np.array(LY), nbIt
```

```python
l0 = .5
t = .001
e = 1e-3
X_uz, Y_uz, n_uz = Uzawa(X0, l0, t, e)
```

```python
XX = [X_uz]
YY = [Y_uz]
forms = ['.']
labels = ['pénalisation extérieure']
print('Le nombre d''itérations prise par la méthode de pénalisation : ',n_uz)
print('Le minimun trouvé est :',X_uz[-1], Y_uz[-1])
Plot(XX, YY, f,forms, labels, "Evolution of the f function during penalisation method")
```
